```python
import requests
import time
import csv
import os
```

```python
#Create a CSV file with the Names and Person IDs of 20,000 actors from the TMDb API.

import os
api_key = os.getenv("API_KEY")


HEADERS = {
    "accept": "application/json",
    "Authorization": api_key
}

BASE_URL_DETAILS = "https://api.themoviedb.org/3/person/{}"
BASE_URL_EXTERNALS = "https://api.themoviedb.org/3/person/{}/external_ids"
CSV_FILE = "actors_freebase_20000.csv"

# Open CSV file for writing
with open(CSV_FILE, mode="w", newline="", encoding="utf-8") as file:
    writer = csv.writer(file)
    writer.writerow(["Person ID", "Name", "Freebase ID"])  # Write header

    for person_id in range(1, 20001):  # Loop from 1 to 20000
        # Get person details (to fetch the name)
        details_response = requests.get(BASE_URL_DETAILS.format(person_id), headers=HEADERS)
        
        if details_response.status_code == 200:
            details_data = details_response.json()
            name = details_data.get("name", "Unknown")
        else:
            print(f"Failed to fetch details for Person ID {person_id}, skipping.")
            continue  # Skip to next iteration

        # Get external IDs (to fetch Freebase ID)
        externals_response = requests.get(BASE_URL_EXTERNALS.format(person_id), headers=HEADERS)
        
        if externals_response.status_code == 200:
            externals_data = externals_response.json()
            freebase_id = externals_data.get("freebase_id", "N/A")
        else:
            freebase_id = "N/A"

        # Save to CSV
        writer.writerow([person_id, name, freebase_id])
        print(f"Saved: {person_id}, {name}, Freebase ID: {freebase_id}")

        time.sleep(0.3)  # Avoid hitting API rate limits

print(f"CSV file '{CSV_FILE}' created successfully!")
```

```python
#take name as a string input and output the person ID

CSV_FILE = "actors_freebase_20000.csv"  # Make sure this is the correct file name

def get_person_id(name):
    with open(CSV_FILE, mode="r", encoding="utf-8") as file:
        reader = csv.reader(file)
        next(reader)  # Skip header row
        
        for row in reader:
            if row[1].strip().lower() == name.strip().lower():  # Case-insensitive match
                return row[0]  # Return Person ID
    
    return "Name not found"

# Example usage
name_input = input("Enter the actor's name: ")
person_id = get_person_id(name_input)

print(f"Person ID for '{name_input}': {person_id}")
```
